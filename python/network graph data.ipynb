```python
link=pd.DataFrame()

for acc in clients:

    df1=df[df["Account_No"]==acc]
    link1=pd.DataFrame()


    link1["source"]=list(df1['receiver_Account'].value_counts().index)
    link1["target"]=acc
    link1["value"]= list(df1['receiver_Account'].value_counts())
    link=pd.concat([link,link1], ignore_index=True)

```
