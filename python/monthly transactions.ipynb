```python
df=pd.read_csv("df18_ts.csv")
date=pd.to_datetime(df.DATE)
df["DATE"]=date
df.head()
```

```python
df.drop(["index","year","month","day"], axis=1, inplace=True)
```

```python
#put date in index
df=df.set_index('DATE')
df.head()
```

```python
clients=list(df.Account_No.unique())
```

```python
clients
```

```python
df18_mo_agg=pd.DataFrame()
for acc in clients:

    df1=df[df["Account_No"]==acc]

    mini=df1.resample('M').min()
    mini.rename(columns={"withd_by_date": "min_mo_withd", "dep_by_date": "min_mo_dep", "Balance_updated": "min_mo_bal"}, inplace=True)

    maxi=df1.resample('M').max()
    maxi.drop(['Account_No'], axis=1, inplace=True)
    maxi.rename(columns={"withd_by_date": "max_mo_withd", "dep_by_date": "max_mo_dep", "Balance_updated": "max_mo_bal"}, inplace=True)


    mean=df1.resample('M').mean()
    mean.rename(columns={"withd_by_date": "mean_mo_withd", "dep_by_date": "mean_mo_dep", "Balance_updated": "mean_mo_bal"}, inplace=True)


    std=df1.resample('M').std()
    std.rename(columns={"withd_by_date": "std_mo_withd", "dep_by_date": "std_mo_dep", "Balance_updated": "std_mo_bal"}, inplace=True)


    median=df1.resample('M').median()
    median.rename(columns={"withd_by_date": "median_mo_withd", "dep_by_date": "median_mo_dep", "Balance_updated": "median_mo_bal"}, inplace=True)

    client=pd.concat([mini,maxi,mean,std,median], axis=1)

    df18_mo_agg=pd.concat([df18_mo_agg,client])
```

```python
df18_mo_agg.reset_index(inplace=True)
```
