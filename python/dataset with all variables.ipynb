```python
# len(clients)
#data of AGGREGATIONS of cash_ounts + dbscan_clusters
data=pd.read_csv("clusters_cash_out.csv")
# data.head()
data['mean']=data['mean'].round(decimals=1)
data['max']=data['max'].round(decimals=1)
data['min']=data['min'].round(decimals=1)
data['median']=data['median'].round(decimals=1)
```

```python
profiles=list(data["dbscan_cluster"].unique())
profiles
```

```python
#prepare the dataset of cashout transactions of accounts belonging to one cluster
def data_by_cluster(data,cash_out,cluster):
    #account numbers of clients of same cluster
    cluster_clients=data[data["dbscan_cluster"]==cluster]["Account_No"].unique()
    #Transactions of clients belonging to a profile
    cluster_transactions=cash_out[cash_out["Account_No"].isin(cluster_clients)]

    #take the transactions of the same month where the profiling is done
    cluster_transactions=cluster_transactions[cluster_transactions["month"]==2]
    return cluster_transactions
```

```python
# cluster_transactions[cluster_transactions["Account_No"]=="1196428'"]["agency"].value_counts()
cluster_transactions=data_by_cluster(data,cash_out,profiles[1])
cluster_transactions
```

```python
#Compute frequency of every agency and amount taken from them per client for the json tree
def data_tree(cluster_transactions):
    stats_withd = cluster_transactions.groupby(['Account_No','agency'])['WITHDRAWAL_AMT'].agg(['sum'])
    stats_withd.columns = [ 'withd_by_agency']
    freq_agency = cluster_transactions.groupby(['Account_No','agency'])['agency'].agg(['count'])
    freq_agency.columns = [ 'freq_by_agency']
    freq_agency['withd_by_agency']=stats_withd['withd_by_agency']
    freq_agency['withd_by_agency']=freq_agency['withd_by_agency'].round(decimals=1)

    # move out account_no and agency from indexes
    freq_agency.reset_index(inplace=True, drop=False)
    freq_agency.reset_index(inplace=True, drop=False)
    return freq_agency
```

```python
#these are the children of salariés 1
def tree_json(tree_data,salaried_name,size_amount):
    tree={"name": salaried_name, "children": []}
    accounts=list(tree_data["Account_No"].unique())
    for index, acc in enumerate(accounts):
        tree["children"].append({"name":acc, "children":[]})
        data_per_account=tree_data[tree_data["Account_No"]==acc]
        data_per_account.reset_index(drop=True, inplace=True)
        for i in range(len(data_per_account)):
            freq=data_per_account["freq_by_agency"][i]
            amount=data_per_account["withd_by_agency"][i]
            agency=data_per_account["agency"][i]
            tree["children"][index]["children"].append({"name":agency,"size_agency":freq,"children":[{"name":amount,size_amount:amount}]})
    return tree
```
